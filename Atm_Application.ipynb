```python
class Atm:
    def __init__(self):
        self.pin=''
        self.balance=0
        self.menu()
        
    def menu(self):
        user_input=input('''How can i help you?
        1. Press 1 for create pin
        2. Press 2 for change pin
        3. Press 3 for check balance
        4. Press 4 for withdraw
        5. Press 5 for exit :- ''')
        
        if user_input=='1':
            self.create_pin()
        elif user_input=='2':
            self.change_pin()
        elif user_input=='3':
            self.check_balance()
        elif user_input=='4':
            self.withdraw()
        else:
            self.exit()
        
    def create_pin(self):
        user_pin=input('enter your pin :-')
        self.pin=user_pin
        
        user_balance=int(input('enter your balance :-'))
        self.balance=user_balance
        print('your pin set successfully')
        self.menu()
        
    def change_pin(self):
        old_pin=input('enter your old pin :-')
        if(old_pin==self.pin):
            new_pin=input('enter your pin :-')
            self.pin=new_pin
            print('your pin set successfully')
            print(self.balance)
            self.menu()
        else:
            print('wrong pin please enter a valid pin')
            self.menu()
            
    def check_balance(self):
        my_pin=input('enter your pin :-')
        if(my_pin==self.pin):
            print(self.balance)
            self.menu()
            
    def withdraw(self):
        my_pin=input('enter your pin :-')
        if(my_pin==self.pin):
            withdraw=int(input('enter your amount :-'))
            if(self.balance>=withdraw):
                print(int(self.balance)-int(withdraw))
                print("collect your amount")
                self.menu()
            else:
                print("Insufficient Amount")
                print(f"Insufficient Amount :{self.withdraw()}")
                
        else:
            print("Please enter correct pin")
            print(self.withdraw())
                
    def exit(self):
        print("Thank you for using")
        self.menu()
        
obj=Atm()

```
